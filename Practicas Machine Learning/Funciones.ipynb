# 1 Funciones

## 1.1 ¿Qué son?

• Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.
• Las funciones van identificadas por un nombre.
• Se debe usar este nombre para invocar (ejecutar) el código de la función.
• Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
• En cada invocación de la función, los parámetros de entrada pueden variar.

## 1.4 Programación de funciones en Python

Las funciones se definen por medio de la palabra clave def.

Formato general:

```
def name(arg1, arg2, ..., argN):
    statements
```

Muy comúnmente:

```
def name(arg1, arg2, ..., argN):
    ...
    return value
```

• Al ejecutarse una sentencia def, se crea un objeto de tipo función y éste se asocia con el nombre especificado para la función.
• Un objeto de tipo función es como cualquier otro tipo de objeto.
• Variables pueden referenciar objetos de tipo función.

````python
# Función que recibe dos números y devuelve su suma.
def sumar(x, y):
    return x + y

# Se asigna la función 'sumar' a otra variable llamada 'f'.
# Ahora 'f' y 'sumar' hacen referencia a la misma función.
f = sumar

# Muestra el identificador (ID) de la referencia almacenada en 'f'.
print(id(f))

# Muestra el identificador (ID) de la función 'sumar'.
# Será el mismo que el de 'f' porque ambas apuntan al mismo objeto.
print(id(sumar))

# Muestra el tipo de 'f'.
# El resultado será <class 'function'>, indicando que es una función.
print(type(f))

# Llama a la función 'sumar' con los valores 5 y 6.
# Devuelve 11.
print(sumar(5, 6))

# Llama a la misma función utilizando la referencia 'f'.
# El resultado también será 11.
print(f(5, 6))
````

````python
# Función que recibe una lista de cadenas y una lista de funciones.
# Aplica cada función a cada cadena y devuelve una nueva lista con los resultados.
def clean_strings(strings, ops):
    result = []  # Lista donde se guardarán las cadenas procesadas.

    # Recorre cada cadena de la lista.
    for value in strings:
        # Aplica todas las funciones de la lista 'ops' en orden.
        for function in ops:
            value = function(value)
        # Guarda la cadena ya transformada.
        result.append(value)

    # Devuelve la lista con todas las cadenas modificadas.
    return result

# Función que agrega el texto "_5" al final de una cadena.
def concat5(x):
    return x + '_5'

# Lista de cadenas con espacios al inicio y/o al final.
strings = [' valencia ', ' barcelona', ' bilbao ']

# Imprime la lista original.
print(strings)

# Lista de funciones que se aplicarán a cada cadena:
# - str.strip: elimina espacios al inicio y al final.
# - str.title: convierte la primera letra de cada palabra en mayúscula.
# - concat5: agrega el sufijo "_5".
operations = [str.strip, str.title, concat5]

# Llama a la función clean_strings, aplica las operaciones
# a cada cadena e imprime el resultado final.
print(clean_strings(strings, operations))
````

## 1.5 Paso de parámetros

• Tipos simples (inmutables) por valor: int, float, string.
  – El efecto es como si se creara una copia dentro de la función, aunque realmente no es esto lo que ocurre.
  – Los cambios dentro de la función no afectan fuera.

````python
# Función que recibe un número como parámetro.
def cambiar_valor(x):
    # Suma 3 al valor recibido.
    # Como los enteros son inmutables, se crea un nuevo objeto con el valor actualizado.
    x += 3

    # Muestra el valor de x dentro de la función.
    print('Dentro de la función:', x)

    # Finaliza la función sin devolver ningún valor.
    return

# Variable con valor inicial.
a = 2

# Se llama a la función pasando el valor de 'a'.
# La variable 'a' no cambia, ya que solo se modifica la copia local 'x'.
cambiar_valor(a)

# Se imprime el valor de 'a' fuera de la función.
# Sigue siendo 2 porque la función no modificó la variable original.
print('Fuera de la función:', a)
````

• Tipos complejos (mutables) por referencia: list, set, dictionary.
  – Dentro de la función se maneja el mismo objeto que se ha pasado desde fuera.
  – Los cambios dentro de la función sí afectan fuera.

````python
# Función que recibe una lista como parámetro.
def anyadir_2_a(x):
    # Agrega el número 4 al final de la lista.
    # Las listas son mutables, por lo que el cambio afecta a la lista original.
    x.append(4)

# Se crea una lista con tres elementos.
lista = [0, 1, 3]

# Se llama a la función pasando la lista.
# La función modifica directamente la lista original.
anyadir_2_a(lista)

# Se imprime la lista después de la modificación.
# El resultado será: [0, 1, 3, 4]
print(lista)
````

````python
# Función que recibe una tupla como parámetro.
def anyadir_2_a(x):
    # Muestra el identificador (ID) del objeto recibido.
    # Si coincide con el de la variable original, significa que ambos hacen referencia
    # al mismo objeto en memoria.
    print(id(x))
    return

# Se crea una tupla con dos elementos.
lista = (0, 1)

# Se imprime el ID de la tupla desde fuera de la función.
print(id(lista))

# Se llama a la función pasando la tupla.
# Como las tuplas son inmutables, no se pueden modificar y simplemente se pasa
# una referencia al mismo objeto.
anyadir_2_a(lista)

# Se imprime la tupla para comprobar que no ha cambiado.
print(lista)
````

• Si quiero modificar un objeto que es de un tipo básico, los cuales se pasan por valor, puedo simplemente devolver el resultado de la función y hacer una asignación:

````python
# Función que recibe un número como parámetro.
def multiplicar_por_2(x):
    # Multiplica el valor recibido por 2.
    # Se crea un nuevo valor y se almacena en la variable local 'x'.
    x = x * 2

    # Devuelve el resultado de la multiplicación.
    return x

# Se crea una variable con el valor 3.
a = 3

# Se llama a la función y el valor devuelto se asigna nuevamente a 'a'.
# De esta forma, 'a' pasa a valer 6.
a = multiplicar_por_2(a)

# Se imprime el valor final de 'a'.
print(a)
````

• Si quiero que no se modifique un objeto de un tipo complejo, los cuales se pasan por referencia, puedo pasar una copia a la función:

````python
# Función que recibe una lista como parámetro.
def anyadir_2_a(x):
    # Agrega el número 2 al final de la lista recibida.
    x.append(2)

    # Muestra la lista modificada dentro de la función.
    print('Dentro de la función:', x)

# Lista original.
lista = [0, 1]

# Se pasa una copia de la lista, por lo que la original no se modifica.
anyadir_2_a(lista.copy())

# Se imprime la lista original para comprobar que sigue igual.
print('Fuera de la función:', lista)
````

````python
# Alternativa para obtener una copia: slicing.
lista = [0, 1]
anyadir_2_a(lista[:])
print('Fuera de la función: ', lista)
````

• Si la lista tiene sublistas anidadas hay que hacer un deep copy

````python
# Se importa el módulo 'copy', que permite crear copias de objetos.
import copy

# Lista original que contiene otra lista anidada.
lista = [2, 4, 16, 32, [34, 10, [5, 5]]]

# Se crea una copia profunda (deep copy) de la lista.
# Esto copia tanto la lista principal como todas las listas internas,
# creando objetos completamente independientes.
copia = copy.deepcopy(lista)

# Se modifica el primer elemento de la copia.
copia[0] = 454

# Se modifica un elemento de la lista anidada dentro de la copia.
copia[4][2][0] = 64

# Se imprime la lista original.
# No cambia porque la copia es completamente independiente.
print(lista)

# Se imprime la copia con las modificaciones realizadas.
print(copia)

# Se muestran los ID de la lista original y de la copia.
# Serán diferentes porque son objetos distintos.
print(f"{id(lista)} - {id(copia)}")

# Se muestran los ID de las listas anidadas.
# También serán diferentes gracias al uso de deepcopy().
print(f"{id(lista[4])} - {id(copia[4])}")
````

• Reasignar un parámetro nunca afecta al objeto de fuera:

````python
# Función que recibe una lista como parámetro.
def funcion_poco_util(x):
    # Muestra el ID del objeto recibido inicialmente.
    # Este ID corresponde a la misma lista que se pasa desde fuera.
    print(id(x))

    # Se reasigna la variable local 'x' a una nueva lista.
    # Esto NO modifica la lista original, solo cambia la referencia local.
    x = [2, 3]

    # Se añade un elemento a la nueva lista local.
    x.append(4)

    # Se muestra el ID de la nueva lista creada dentro de la función.
    print(id(x))

# Lista original.
lista = [0, 1]

# Se imprime el ID de la lista original.
print(id(lista))

# Se llama a la función pasando la lista.
funcion_poco_util(lista)

# Se imprime la lista original después de la función.
# No cambia porque la función solo modificó una copia local.
print(lista)
````

````python
# Función que recibe una tupla como parámetro.
def modif_tupla(a):
    # Muestra el ID del objeto recibido.
    print(id(a))

# Se crea una tupla.
t = (1, 2)

# Se muestra el ID de la tupla original.
print(id(t))

# Se llama a la función pasando la tupla.
# La tupla no puede modificarse porque es inmutable.
modif_tupla(t)

# Se imprime la tupla original.
# Permanece igual porque no ha sido modificada.
print(t)
````

## 1.6 Argumentos arbitrarios

• No se sabe a priori cuantos elementos se reciben.
• Se reciben como una tupla.
• Los parámetros se especifican con el símbolo '*'.

````python
# Función que recibe un número variable de argumentos.
# El operador * empaqueta los argumentos en una tupla llamada 'names'.
def saludar(*names):

    # Muestra el tipo de dato de 'names'.
    # Siempre será una tupla.
    print(type(names))

    # Recorre cada nombre dentro de la tupla.
    for name in names:
        # Imprime un saludo para cada nombre.
        print("Hola " + name)

# Llamada sin argumentos.
# 'names' será una tupla vacía.
saludar()

# Llamada con varios argumentos.
# Todos los valores se agrupan en la tupla 'names'.
saludar("Manolo", "Pepe", "Luis", "Alex", "Juan")
````

````python
# Función que recibe un número variable de argumentos con nombre.
# El operador ** empaqueta los argumentos en un diccionario llamado 'kwargs'.
def consulta_sql(**kwargs):

    # Imprime el diccionario con los parámetros recibidos.
    print(kwargs)

    # Se inicia la construcción de la consulta SQL.
    sql = "SELECT * FROM bicis WHERE "

    # Se crea una lista de condiciones a partir del diccionario.
    # Cada elemento tiene la forma "clave=valor".
    args = [f"{key}={value}" for key, value in kwargs.items()]

    # Se unen todas las condiciones con " AND ".
    sql += " AND ".join(args)

    # Se devuelve la consulta SQL completa.
    return sql

# Llamada a la función con argumentos nombrados.
# Estos se convierten en un diccionario dentro de la función.
print(consulta_sql(station='alameda', free=10))
````

## 1.7 Recursividad

• Funciones que se llaman a sí mismas

````python
# Función que calcula el factorial de un número usando recursividad.
def factorial(n):
    # Caso base: el factorial de 0 es 1.
    if n == 0:
        return 1
    else:
        # Llamada recursiva: n * factorial(n-1)
        return n * factorial(n - 1)

# Imprime el factorial de 10 usando la versión recursiva.
print(factorial(10))


# Función que calcula el factorial de un número usando un bucle.
def fact(n):
    # Inicializa el acumulador en 1.
    a = 1

    # Recorre desde 1 hasta n (incluido).
    for i in range(1, n + 1):
        # Multiplica acumulativamente.
        a *= i

    # Devuelve el resultado final.
    return a

# Imprime el factorial de 10 usando la versión iterativa.
print(fact(10))
````

````python
# Función factorial definida como una lambda (función anónima).
# Usa una condición ternaria:
# - Si x == 0, devuelve 1 (caso base).
# - Si no, devuelve x * fact(x-1) (recursividad).
fact = lambda x: 1 if x == 0 else x * fact(x - 1)

# Llama a la función para calcular el factorial de 10.
fact(10)
````

## 1.8 Funciones Lambda

• Es posible definir funciones anónimas (sin nombre).
• Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia def no puede (por ejemplo, dentro de una lista o como parámetro de una función).
• Útiles cuando se quiere pasar una función como argumento a otra.

Formato general:

```
lambda <lista de argumentos> : <valor a retornar>
```

• Importante: <valor a retornar> no es un conjunto de instrucciones. Es simplemente una expresión return que omite esta palabra.
• Las funciones definidas con def son más generales:
  – Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con def, pero no viceversa.

````python
# Ejemplo de función normal que suma tres números.
def suma(x, y, z):
    # Devuelve la suma de los tres parámetros.
    return x + y + z

# Llamada a la función suma.
print(suma(1, 2, 3))


# Ejemplo de la misma operación usando una función lambda.
# Las funciones lambda son funciones anónimas y más compactas.
suma2 = lambda x, y, z: x + y + z

# Llamada a la función lambda.
print(suma2(1, 2, 3))
````

````python
# El método sort() puede recibir un parámetro opcional llamado key.
# Este parámetro es una función que se aplica a cada elemento antes de comparar.

# Ejemplo: lista de ciudades.
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# Ordenación por defecto (alfabética).
cities.sort()
print(cities)

# Ordenación usando una lambda que devuelve la longitud de cada string.
cities.sort(key=lambda x: len(x))
print(cities)

# Ordenación usando directamente la función len (más simple).
cities.sort(key=len)
print(cities)

# Función personalizada que devuelve la longitud de una cadena.
def longitud(x):
    return len(x)

# Ordenación usando la función definida por el usuario.
cities.sort(key=longitud)
print(cities)
````

````python
# Mismo ejemplo sin usar lambda.
# Ordenación de una lista de ciudades según la longitud de sus nombres.

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# Función que devuelve el número de caracteres de una cadena.
def num_caracteres(x):
    return len(x)

# Ordena la lista usando la función como clave de ordenación.
cities.sort(key=num_caracteres)

# Imprime la lista ya ordenada por longitud de los nombres.
print(cities)
````

• Nos podemos guardar una referencia para utilización repetida.

````python
# Lista de ciudades.
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

# Función lambda que devuelve la longitud de una cadena.
f = lambda x: len(x)

# Recorre cada ciudad de la lista.
for s in cities:
    # Imprime la longitud de cada ciudad usando la función lambda.
    print(f(s))
````

• Las expresiones lambda pueden formar parte de una lista.

````python
# Lista de funciones lambda que calculan potencias de x.
pows = [
    lambda x: x ** 0,
    lambda x: x ** 1,
    lambda x: x ** 2,
    lambda x: x ** 3,
    lambda x: x ** 4
]

# Recorre cada función de la lista.
for f in pows:
    # Llama a cada función con el valor 2 e imprime el resultado.
    print(f(2))
````

• Diccionarios se pueden utilizar como 'switch' en otros lenguajes
• Ideal para escoger entre varias opciones

````python
# Función que simula un "switch" usando un diccionario de funciones.
def switch_dict(operator, x, y):
    # Diccionario que asocia cada operación con una función lambda.
    return {
        'add': lambda: x + y,  # suma
        'sub': lambda: x - y,  # resta
        'mul': lambda: x * y,  # multiplicación
        'div': lambda: x / y,  # división
    # get() busca la clave 'operator'.
    # Si no existe, devuelve lambda: None.
    }.get(operator, lambda: None)()

# Llamadas a la función con distintos operadores.

print(switch_dict('add', 2, 2))   # 4
print(switch_dict('div', 10, 2))  # 5.0
print(switch_dict('mod', 17, 3))  # None (operador no definido)
````

## 1.9 Scopes

• La localización de una asignación a una variable en el código determina desde donde puedes acceder a esa variable.
• Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
• El alcance de la visibilidad de una variable determina su scope.
• El término scope hace referencia a un espacio de nombres (namespace). Por ejemplo, una función establece su propio namespace.

````python
# Variable global (no utilizada dentro de la función en este caso).
X = 10

# Función principal.
def func():

    # Variable local a la función 'func'.
    # No modifica la variable global X, es una variable distinta.
    X = 20

    # Función interna (anidada) dentro de 'func'.
    def func_int():

        # Variable local a 'func_int'.
        # Tampoco afecta a las otras variables X.
        X = 30

        # Imprime la variable local de esta función interna.
        print(X)

    # Llamada a la función interna.
    func_int()

# Llamada a la función principal.
func()
````

**Regla LEGB** Dada una función F, tenemos los siguientes scopes:

• Local: variables locales a F.
• Enclosing: variables localizadas en funciones que contienen a (o por encima de) F.
• Global: variables definidas en el módulo (fichero) que no están contenidas en ninguna función. Este scope no abarca más de un fichero.
• Built-ins: proporcionadas por el lenguaje.

La resolución de nombres ocurre de abajo a arriba.

````python
# Variable global.
X = 99

# Función que recibe un parámetro Y.
def func(Y):
    # Variable local Z, calculada usando la variable global X y el parámetro Y.
    Z = X + Y
    return Z

# Llamada a la función con el valor 1.
# Devuelve 99 + 1 = 100.
print(func(1))
````

````python
# Variable global X.
X = 1

# Función que define una variable local X.
def func():
    # Esta X es local a la función y no modifica la global.
    X = 2

# Llamada a la función.
func()

# Se imprime la variable global X.
# Sigue valiendo 1 porque la variable dentro de la función era local.
print(X)
````

• La sentencia global nos permite modificar una variable global desde dentro de una función.

````python
# Variable global X.
X = 1

# Función que modifica la variable global.
def func():
    # Indica que se va a usar la variable global X,
    # no una variable local.
    global X

    # Modifica la variable global X.
    X = 2

# Llamada a la función.
func()

# Se imprime el valor de X después de la modificación.
# Ahora vale 2 porque fue cambiada dentro de la función.
print(X)
````

• No hay necesidad de usar global para referenciar variables. Sólo es necesario para modificarlas.

````python
# Se crean dos variables globales.
y, z = 1, 2

# Función que crea/modifica una variable global.
def todas_globales():
    # Se declara que 'x' será global.
    global x

    # Se calcula usando variables globales y se guarda en 'x'.
    x = y + z  # Las tres variables terminan siendo globales.

# Llamada a la función.
todas_globales()

# Imprime la variable global creada dentro de la función.
print(x)

# Imprime las variables globales originales.
print(y)
print(z)
````

````python
# Variable global.
X = 77

# Función externa.
def f1():
    # Variable local de f1 (enclosing scope para f2).
    X = 88

    # Función interna.
    def f2():
        # Busca X en el ámbito más cercano (f1), no en el global.
        print(X)

    # Llamada a la función interna.
    f2()

# Llamada a la función externa.
f1()
````

**Closures**

• Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan existiendo o no.

````python
# Función externa.
def f1():
    # Variable local de f1 (enclosing scope para f2).
    X = 88

    # Función interna.
    def f2():
        # Accede a X desde el ámbito envolvente (f1).
        print(X)

    # Devuelve la función f2 sin ejecutarla.
    return f2

# Se ejecuta f1, que devuelve la función f2.
accion = f1()

# Se llama a la función devuelta.
accion()
````

• Sentencia nonlocal para modificar variables que no son locales, pero tampoco globales.

````python
# Función externa.
def f1():
    # Variable en el ámbito envolvente (enclosing scope).
    contador = 10

    # Función interna.
    def f2():
        # Indica que 'contador' no es local,
        # sino que pertenece a la función envolvente f1.
        nonlocal contador

        # Se incrementa el valor de contador.
        contador += 1

        # Se imprime el valor actualizado.
        print(contador)

    # Se devuelve la función interna sin ejecutarla.
    return f2

# Se obtiene la función f2 con su estado (closure).
accion = f1()

# Cada llamada modifica y recuerda el valor de 'contador'.
accion()  # 11
accion()  # 12
accion()  # 13
````

## 1.10 Ejercicios

1. Escribe una función que reciba como entrada una lista con números y devuelva como resultado una lista con los cuadrados de los números contenidos en la lista de entrada.
2. Escribe una función que reciba números como entrada y devuelva la suma de los mismos. La función debe ser capaz de recibir una cantidad indeterminada de números. La función no debe recibir directamente ningún objeto complejo (lista, conjunto, etc.).
3. Escribe una función que reciba un string como entrada y devuelva el string al revés. Ejemplo: si el string de entrada es 'hola', el resultado será 'aloh'.
4. Escribe una función lambda que, al igual que la función desarrollada en el ejercicio anterior, invierta el string recibido como parámetro. Ejemplo: si el string de entrada es 'hola', el resultado será 'aloh'.
5. Escribe una función que compruebe si un número se encuentra dentro de un rango específico.
6. Escribe una función que reciba un número entero positivo como parámetro y devuelva una lista que contenga los 5 primeros múltiplos de dicho número. Por ejemplo, si la función recibe el número 3, devolverá la lista [3, 6, 9, 12, 15]. Si la función recibe un parámetro incorrecto (por ejemplo, un múmero menor o igual a cero), mostrará un mensaje de error por pantalla y devolverá una lista vacía.
7. Escribe una función que reciba una lista como parámetro y compruebe si la lista tiene duplicados. La función devolverá True si la lista tiene duplicados y False si no los tiene.
8. Escribe una función lambda que, al igual que la función desarrollada en el ejercicio anterior, reciba una lista como parámetro y compruebe si la lista tiene duplicados. La función devolverá True si la lista tiene duplicados y False si no los tiene.
9. Escribe una función que compruebe si un string dado es un palíndromo. Un palíndromo es una secuencia de caracteres que se lee igual de izquierda a derecha que de derecha a izquierda. Por ejemplo, la función devolverá True si recibe el string "reconocer" y False si recibe el string "python".

````python
# Ejercicio 1
def cuadrados_lista(lista):
    # Devuelve una lista con los cuadrados de cada número
    return [x ** 2 for x in lista]

# Ejemplo de uso
print(cuadrados_lista([1, 2, 3, 4]))
````

````python
# Ejercicio 2
def suma(*numeros):
    # Suma una cantidad indeterminada de números
    return sum(numeros)

# Ejemplo de uso
print(suma(1, 2, 3, 4))
````

````python
# Ejercicio 3
def invertir_string(s):
    # Devuelve el string invertido
    return s[::-1]

# Ejemplo de uso
print(invertir_string("hola"))
````

````python
# Ejercicio 4
# Lambda que invierte un string
invertir = lambda s: s[::-1]

# Ejemplo de uso
print(invertir("aguila"))
````

````python
# Ejercicio 5
def en_rango(n, minimo, maximo):
    # Comprueba si un número está dentro de un rango
    return minimo <= n <= maximo

# Ejemplo de uso
print(en_rango(5, 1, 10))
````

````python
# Ejercicio 6
def primeros_multiplos(n):
    # Si el número no es válido, devuelve lista vacía
    if n <= 0:
        print("Error: número inválido")
        return []

    # Devuelve los 5 primeros múltiplos del número
    return [n * i for i in range(1, 6)]

# Ejemplo de uso
print(primeros_multiplos(3))
````

````python
# Ejercicio 7
def tiene_duplicados(lista):
    # Devuelve True si hay elementos repetidos
    return len(lista) != len(set(lista))

# Ejemplo de uso
print(tiene_duplicados([1, 2, 3, 3]))
````

````python
# Ejercicio 8
# Lambda que detecta duplicados en una lista
tiene_duplicados_lambda = lambda lista: len(lista) != len(set(lista))

# Ejemplo de uso
print(tiene_duplicados_lambda([1, 2, 3, 4]))
````

````python
# Ejercicio 9
def es_palindromo(texto):
    # Comprueba si el texto es igual al revés
    return texto == texto[::-1]

# Ejemplo de uso
print(es_palindromo("reconocer"))
print(es_palindromo("python"))
````
