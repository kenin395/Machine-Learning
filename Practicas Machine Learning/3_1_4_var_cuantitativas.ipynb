```python
# Creación aleatoria de datos
import numpy as np
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5,
-28, -50, 60, -100, -10, -11, 13, 19, 1, 9])
print(X)
```

```python
# Importar el módulo de estadísticas de SciPy
from scipy import stats

# Calcular la media (promedio) de los datos almacenados en X
media = np.mean(X)

# Calcular la mediana (valor central de los datos)
mediana = np.median(X)

# Calcular la moda (valor que más se repite)
# stats.mode() devuelve la moda y su frecuencia
moda, _ = stats.mode(X)

# Mostrar el valor de la media
print('Media:', media)

# Mostrar el valor de la mediana
print('Mediana:', mediana)

# Mostrar el valor de la moda
print('Moda:', moda)
```

```python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Calcular el primer cuartil (25 % de los datos)
Q1 = np.percentile(X, 25)

# Calcular el tercer cuartil (75 % de los datos)
Q3 = np.percentile(X, 75)

# Calcular el Rango Intercuartílico (RIC)
# El RIC mide la dispersión de la mitad central de los datos
RIC = Q3 - Q1

# Calcular el límite inferior para detectar posibles valores atípicos
lim_inf = Q1 - 1.5 * RIC

# Calcular el límite superior para detectar posibles valores atípicos
lim_sup = Q3 + 1.5 * RIC

# Mostrar el valor del Rango Intercuartílico
print('RIC:', RIC)

# Mostrar el límite inferior
print('Límite inferior:', lim_inf)

# Mostrar el límite superior
print('Límite superior:', lim_sup)
```

```python
 # Dibuja el diagrama de box and whiskers y determina visualmente el número de␣outliers
plt.boxplot(X)
plt.grid()
plt.title("Diagrama de caja y bigotes")
plt.show()
```

```python
# Medidas de dispersión

# Calcular la desviación estándar (o desviación típica) de los datos
desvest = np.std(X)

# Mostrar el valor de la desviación estándar
print('Desviación típica:', desvest)

# Calcular y mostrar la varianza
# La varianza es el cuadrado de la desviación estándar
print('Varianza:', desvest ** 2)
```

```python
# Medidas de distribución

# Calcular el coeficiente de asimetría de los datos
# Indica si la distribución está inclinada hacia la izquierda o la derecha
asimetria = stats.skew(X)

# Calcular la curtosis de los datos
# fisher=True hace que una distribución normal tenga curtosis igual a 0
curtosis = stats.kurtosis(X, fisher=True)

# Mostrar la asimetría redondeada a 4 decimales
print('Asimetría:', np.round(asimetria, 4))

# Mostrar la curtosis redondeada a 4 decimales
print('Curtosis:', np.round(curtosis, 4))

# Determinar el tipo de distribución según la curtosis
if curtosis > 0:
    print('Distribución Leptocúrtica')
elif curtosis < 0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')
```

```python
# Importar el conjunto de datos Iris desde scikit-learn
from sklearn import datasets

# Importar pprint para mostrar la información de forma organizada
from pprint import pprint

# Cargar el conjunto de datos Iris
iris = datasets.load_iris()

# Mostrar toda la información del conjunto de datos
pprint(iris)

# Guardar las características (variables) en X
X = iris.data

# Guardar las clases (especies de flores) en y
y = iris.target
```

¿Qué tipo de asimetría se observa en la variable "longitud del pétalo"?

```python
# Importar la función para calcular la asimetría
from scipy import stats

# Obtener la columna correspondiente a la longitud del pétalo
# (columna 2 del conjunto de datos Iris)
longitud_petalo = X[:, 2]

# Calcular la asimetría
asimetria = stats.skew(longitud_petalo)

# Mostrar el valor de la asimetría
print("Asimetría:", round(asimetria, 4))

# Determinar el tipo de asimetría
if asimetria > 0:
    print("La longitud del pétalo presenta una asimetría positiva.")
elif asimetria < 0:
    print("La longitud del pétalo presenta una asimetría negativa.")
else:
    print("La longitud del pétalo presenta una distribución simétrica.")
```



¿A qué tipo de distribución se ajusta la variable "ancho del sépalo", según su curtosis?

```python
# Obtener la columna correspondiente al ancho del sépalo
# (columna 1 del conjunto de datos Iris)
ancho_sepalo = X[:, 1]

# Calcular la curtosis
curtosis = stats.kurtosis(ancho_sepalo, fisher=True)

# Mostrar el valor de la curtosis
print("Curtosis:", round(curtosis, 4))

# Determinar el tipo de distribución
if curtosis > 0:
    print("El ancho del sépalo tiene una distribución Leptocúrtica.")
elif curtosis < 0:
    print("El ancho del sépalo tiene una distribución Platicúrtica.")
else:
    print("El ancho del sépalo tiene una distribución Mesocúrtica.")
```
