```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
# Crear una lista vacía para almacenar los niveles de estudios
nivel_estudios = []

# Agregar 110 registros con el valor "Secundaria"
nivel_estudios.extend(['Secundaria'] * 110)

# Agregar 172 registros con el valor "Bachillerato"
nivel_estudios.extend(['Bachillerato'] * 172)

# Agregar 145 registros con el valor "Grado"
nivel_estudios.extend(['Grado'] * 145)

# Agregar 52 registros con el valor "Master"
nivel_estudios.extend(['Master'] * 52)

# Agregar 21 registros con el valor "Doctorado"
nivel_estudios.extend(['Doctorado'] * 21)

# Mezclar aleatoriamente el orden de todos los niveles de estudio
# para que no queden agrupados por categoría
nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Obtener los valores únicos de la lista "nivel_estudios"
# y contar cuántas veces aparece cada uno
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)

print(estudios, frecuencia)
```

```python
# Asignar una etiqueta al eje Y (vertical)
plt.ylabel('Frecuencia absoluta')

# Asignar una etiqueta al eje X (horizontal)
plt.xlabel('Nivel de estudios')

# Agregar un título al gráfico
plt.title('Diagrama de barras')

# Crear un gráfico de barras
# "estudios" contiene las categorías del eje X
# "frecuencia" contiene la cantidad de registros de cada categoría
plt.bar(estudios, frecuencia)

# Mostrar el gráfico en pantalla
plt.show()
```

```python
# DIAGRAMA DE SECTORES (GRÁFICO CIRCULAR)

# Definir los colores que tendrá cada sector del gráfico
colores = {'red', 'blue', 'orange', 'green', 'cyan'}

# Crear el gráfico circular
# frecuencia: tamaño de cada sector
# labels: nombres de cada categoría
# colors: colores asignados a cada sector
# autopct: muestra el porcentaje con dos decimales

plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')

plt.title('Diagrama de sectores')


plt.show()
```

1. Crear el DataFrame

```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con los niveles de estudio y sus frecuencias
df = pd.DataFrame({
    'Nivel de estudios': ['Secundaria', 'Bachillerato', 'Grado', 'Master', 'Doctorado'],
    'Frecuencia': [110, 172, 145, 52, 21]
})

print(df)
```

2. Diagrama de barras

```python
# Crear un gráfico de barras utilizando el DataFrame
plt.bar(df['Nivel de estudios'], df['Frecuencia'])

# Agregar una etiqueta al eje X
plt.xlabel('Nivel de estudios')

# Agregar una etiqueta al eje Y
plt.ylabel('Frecuencia absoluta')

plt.title('Diagrama de barras')

plt.show()
```

3. Diagrama de sectores (pastel)

```python
# Definir los colores para cada sector del gráfico
colores = ['red', 'blue', 'orange', 'green', 'cyan']

# Crear el gráfico de sectores utilizando el DataFrame
plt.pie(
    df['Frecuencia'],
    labels=df['Nivel de estudios'],
    colors=colores,
    autopct='%1.2f%%'
)

plt.title('Diagrama de sectores')

plt.show()
```
