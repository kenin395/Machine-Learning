## 03
- Resolución de dos sistemas de ecuaciones
    - $ A : \begin{cases} -9x +  4y = 20 \\ -7y + 16x = 80                         \end{cases}$
    
    - $ B : \begin{cases} x - 2y + 3z = 7 \\ 2x + y + z = 4 \\ -3x + 2y -2z = -10 \end{cases}$
- Verifica por código los resultados

```python
# Importa la biblioteca matemática para usar isclose()
import math

# Importa NumPy con el alias np
import numpy as np

# Importa el módulo de álgebra lineal de NumPy
import numpy.linalg as linalg

# ==========================
# PRIMER SISTEMA DE ECUACIONES
# ==========================

# Matriz de coeficientes
a = np.array([
    [-9, 4],
    [16, -7]
])

# Vector de términos independientes
b = np.array([20, 80])

# Muestra la matriz y el vector
print(a)
print(b)

# Resuelve el sistema Ax = b
x, y = linalg.solve(a, b)

# Imprime la solución
print(f'x = {x}, y = {y}')

# Verifica que la solución satisface las ecuaciones
print(math.isclose(-9*x + 4*y, 20))
print(math.isclose(16*x - 7*y, 80))

# ==========================
# SEGUNDO SISTEMA DE ECUACIONES
# ==========================

# Matriz de coeficientes
A2 = np.array([
    [1, -2, 3],
    [2, 1, 1],
    [-3, 2, -2]
])

# Vector de términos independientes
b2 = np.array([7, 4, -10])

# Resuelve el sistema Ax = b
x, y, z = linalg.solve(A2, b2)

# Imprime la solución
print(f'x = {x}, y = {y}, z = {z}')

# Comprueba que la solución es correcta
print(math.isclose(x - 2*y + 3*z, 7))
print(math.isclose(2*x + y + z, 4))
print(math.isclose(-3*x + 2*y - 2*z, -10))
```

```python
# Función que elimina elementos duplicados de una lista
# manteniendo el orden en el que aparecen.
def remove_duplicates(lista):
    
    # Crea una lista vacía para guardar los elementos únicos.
    unicos = []
    
    # Recorre cada elemento de la lista original.
    for elemento in lista:
        
        # Si el elemento todavía no está en la lista de únicos,
        # lo añade.
        if elemento not in unicos:
            unicos.append(elemento)
    
    # Convierte la lista de elementos únicos en un DataFrame
    # y lo devuelve.
    return pd.DataFrame(unicos)
```

```python
# Función que elimina elementos repetidos de una lista
def remove_duplicates(lista):
    
    # Convierte la lista en un conjunto (set).
    # Un set elimina automáticamente los valores duplicados.
    # IMPORTANTE: no garantiza mantener el orden original.
    noduplis = list(set(lista))
    
    # Convierte la lista sin duplicados en un DataFrame.
    # La columna se llama "elementos".
    df = pd.DataFrame(noduplis, columns=['elementos'])
    
    # Devuelve el DataFrame creado.
    return df
```

## 07 - 1pt
- Escribir un generador `frange` que simule el comportamiento de la función `arange` de NumPy
- Esta función cada vez que se invoque debe devolver valores espaciados uniformemente a partir de un valor inicial. Se debe proveer a dicha función un valor inicial para la secuencia de números y un valor para el step
- Debe soportar valores númericos en coma flotante
- La función potencialmente se puede llamar un número infinito de veces
- Usando plain Python, sin ninguna librería

```python
# Importa la librería matemática para usar math.isclose()
import math

# Valor inicial de la secuencia
init = 10.6

# Incremento que tendrá cada número
step = 0.8

# Crea el generador llamando a la función frange()
g = frange(init, step)

# Cantidad de valores que se quieren obtener
iters = 10

# Obtiene los primeros 10 valores del generador
numbers = [next(g) for n in range(iters)]

# Comprueba que cada valor generado sea igual a:
# inicio + (incremento * posición)
for i, n in enumerate(numbers):
    assert(math.isclose(n, init + (step * i)))

# Muestra la lista de números generados
print(numbers)
```

```python
def frange(init, step):
    # Guarda el valor actual de la secuencia
    actual = init
    
    # Genera valores infinitamente
    while True:
        yield actual
        
        # Incrementa el valor para la siguiente iteración
        actual += step
```

- Crear una función que reciba una Serie de Pandas y devuelva esa misma serie, reemplazando los espacios en blanco por el carácter menos frecuente en dicha serie.

```python
# Crea una Serie de pandas a partir de una lista de caracteres.
# La cadena contiene espacios en blanco que deben ser limpiados.
serie = pd.Series(list('bba cabc faabba aacbbfe'))

# Define el resultado esperado después de limpiar los espacios.
# Los espacios deben eliminarse, dejando solamente los caracteres.
expected_output = pd.Series(list('bbaecabcefaabbaeaacbbfe'))

# Llama a la función clean_whitespace_replace().
# Esta función debe eliminar o reemplazar los espacios en blanco de la Serie.
output = clean_whitespace_replace(serie)

# Comprueba que el resultado obtenido sea exactamente igual al esperado.
# Si no coincide, se genera un AssertionError.
assert(expected_output.equals(output))
```

```python
def clean_whitespace_replace(serie):
    # Une todos los elementos de la Serie en un solo texto,
    # elimina los espacios y vuelve a crear una Serie de caracteres.
    texto = ''.join(serie).replace(' ', '')
    
    return pd.Series(list(texto))
```

```python
from collections import Counter

# Función para limpiar/reemplazar espacios en una Serie
def clean_whitespace_replace(seri):
    
    # Une todos los elementos de la Serie en un solo texto.
    # dropna() elimina valores vacíos (NaN).
    # astype(str) convierte todos los elementos a texto.
    texto = ''.join(seri.dropna().astype(str))
    
    # Cuenta la frecuencia de cada carácter del texto.
    # Ignora los caracteres vacíos.
    frecu = Counter(i for i in texto if i != '')
    
    # Busca el carácter que aparece menos veces.
    mefrecu = min(frecu, key=frecu.get)
    
    # Reemplaza cada espacio ' ' por el carácter menos frecuente.
    seridif = seri.astype(str).apply(lambda s: s.replace(' ', mefrecu))
    
    # Devuelve la Serie modificada.
    return seridif
```
