## 01 - 3.3pt
- Genera una matriz con valores aleatorios de 0 a 100 de dimensiones 16x16
- Calcula los siguentes estadísticos en la matriz anterior, por fila (el resultado de cada uno debe ser un array de dimensiones 1x16)
  - Suma de los elementos de cada fila
  - Media de los elementos de cada fila
  - Valores mínimos de cada fila
  - Valores máximos de cada fila

```python
# matriz de 16 x 16 con enteros con valores aleatorios 
matriz = np.random.randint(0,100, size=(16,16))
display(matriz)

#suma de cada fila
suma_filas = matriz.sum(axis=1)
print("sumas por filas;", suma_filas)

#calculacion de la media
#sumar todos los valores del conjunto de datos y dividir el resultado entre el número total de elementos
media_filas = matriz.mean(axis=1)
print("media por filas;", media_filas)

#calcular valores minimos
min_filas = matriz.min(axis=1)
print("min por filas;", min_filas)

#calcular los valores maximos
max_filas = matriz.max(axis=1)
print("max por filas;", max_filas)
```

## 03 - 3.3pt

- Generar un dataframe de 3x4 dimensiones con valores númericos aleatorios [0,10]
- Darle un índice semántico alfanumérico
- Aplicarle una función lambda que obtenga la diferencia entre el máximo y mínimo de cada fila

```python

# Generamos un DataFrame de 3 filas x 4 columnas
# con números enteros aleatorios entre 0 y 10
df = pd.DataFrame(
    np.random.randint(0, 11, size=(3, 4)),
    index=["Fila_A", "Fila_B", "Fila_C"],      
    columns=["Col1", "Col2", "Col3", "Col4"]   
)

# Mostramos el DataFrame original
print("DataFrame original:")
print(df)

# Aplicamos una función lambda a cada fila (axis=1)
# La función calcula la diferencia entre el valor máximo y el mínimo de la fila
diferencia = df.apply(lambda fila: fila.max() - fila.min(), axis=1)

# Mostramos el resultado
print("\nDiferencia entre el máximo y el mínimo de cada fila:")
print(diferencia)
```

## 07 - 3.3pt
- La serie de Fibonacci es una sucesión de números, en la cual cada número es la suma de los dos anteriores. Los dos primeros son siempre 0 y 1.
  - $F_0 = 0$
  - $F_1 = 1$
  - $F_n = F_{n-1} + F_{n-2}$
- Crear un generador infinito de números de Fibonacci.

```python
# Definimos una función generadora
def fibonacci():
    # Los dos primeros números de la serie
    a = 0
    b = 1

    # Bucle infinito
    while True:
        # Devuelve el valor actual
        yield a

    #convierte una función en un generador.

        # Actualiza los valores para el siguiente número
        a, b = b, a + b


# Crear el generador
serie = fibonacci()

# Mostrar los primeros 15 números de Fibonacci
print("Primeros 15 números de Fibonacci:")

for i in range(15):
    print(next(serie))
```

## 07 - 3.3pt
- Crear una lista de meses de Enero a Diciembre (eje X)
- Generar datos aleatorios para el eje Y, hacerlo 3 veces distintas (Y0, Y1 e Y2)
- Representar las 3 secuencias de datos aleatorios en una sola figura usando matplotlib
- Añadir una leyenda para poder identificar cada secuencia
- Nombrar las 3 secuencias de la siguiente manera ("Facebook", "Twitter", "Instagram")
- Añadir un título y nombres a los dos ejes
- Añadir una anotación (texto y flecha) a uno de los gráficos

```python
# Your solution HERE !!!!
# Lista de meses de Enero a Diciembre eje X
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
         'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

# Tres series aleatorias de datos je Y una por red social
y0 = np.random.randint(50, 200, size=12)
y1 = np.random.randint(50, 200, size=12)
y2 = np.random.randint(50, 200, size=12)

# Creamos la figura y dibujamos las 3 líneas en el mismo gráfico
fig, ax = plt.subplots()
ax.plot(meses, y0, 'b-o', label='Facebook')
ax.plot(meses, y1, 'g-s', label='Twitter')
ax.plot(meses, y2, 'r-^', label='Instagram')

# Leyenda para identificar cada secuencia
ax.legend(loc='best')
# Título y nombres de ejes
ax.set_title('Actividad mensual en redes sociales')
ax.set_xlabel('Mes')
ax.set_ylabel('Usuarios activos (miles)')
# Rotamos etiquetas del eje X para que se lean mejor
plt.xticks(rotation=45)

# Anotación con texto y flecha en un punto del gráfico de Facebook
ax.annotate('Pico de actividad',
            xy=(meses[5], y0[5]),
            xytext=(meses[2], y0[5] + 30),
            arrowprops=dict(facecolor='red', shrink=0.05))

plt.tight_layout()
plt.show()
```
