# 1 Acceso a ficheros

## 1.1 Ficheros
`File` es un tipo de objeto predefinido en Python (built-in). Permite acceder a ficheros desde programas Python.
* Los ficheros son de un tipo especial: Son built-in, pero no son ni números, ni secuencias, ni mappings. Tampoco responden a operadores en expresiones.
* La función `open` permite crear objetos de tipo fichero. 
Formato general para abrir un fichero: `afile = open(filename, mode)`
`mode` es opcional. Por defecto, los ficheros se abren en modo lectura.
* Los datos leídos de un fichero siempre se obtienen en formato string. Lo mismo ocurre con escritura.
* Los ficheros se deben cerrar invocando `close` (liberación de recursos).

```python
# Lectura desde fichero usando método 'read'. 
# Devuelve todo el contenido del fichero.
mi_fichero = open('res/multiple_lines.txt')
print(mi_fichero.read())
mi_fichero.close()
```

```python
# Lectura línea a línea a través del bucle 'for'.
mi_fichero = open('res/multiple_lines.txt')
for linea in mi_fichero:
    print (f"{linea}")
mi_fichero.close()
```

```python
# Leer de un archivo a una lista.
mi_fichero = open('res/multiple_lines.txt')
lineas = mi_fichero.readlines()
mi_fichero.close()
print(lineas)
```

Resolución de paths independiente de plataforma

```python
# Importamos os para el manejo de rutas independiente del sistema operativo
import os
ruta = os.path.join("res", "multiple_lines.txt")
print(ruta)
```

```python
# Desempaquetado de una lista para formar la ruta
dataset_file_path = ["res", "multiple_lines.txt"]
ruta = os.path.join(*dataset_file_path)
print(ruta)
```

`close` automático con sentencia `with`. Esta es la forma habitual de leer de fichero en Python, context manager.

```python
# Lectura recomendada usando un context manager (with)
with open(ruta) as mi_fichero:
    for linea in mi_fichero:
        print(linea, end='')
```

Se puede abrir varios ficheros en un mismo `with`.

```python
# Abrir varios ficheros en el mismo bloque with
ruta1 = os.path.join("res", "one_line.txt")
ruta2 = os.path.join("res", "multiple_lines.txt")

with open(ruta1) as fichero1, open(ruta2) as fichero2:
    print(fichero1.readlines())
    print(fichero2.readlines())
```

## 1.2 Modos de acceso
Al crear un objeto de tipo File se puede espeficiar el modo de acceso (lectura/escritura).

| Modo | Acceso Descripción |
|---|---|
| r | Solo Lectura |
| w | Solo Escritura (Borra si el archivo ya existe) |
| x | Solo Escritura (Falla si el archivo ya existe) |
| a | Crea Fichero (Si existe lo abre y se añade al final) |
| r+ | Lectura y Escritura |
| b | Se puede añadir a otros modos para acceso binario |
| t | Acceso para archivos de texto (default) |

## 1.3 Acceso para escritura

```python
# Definición de función para crear una lista de cadenas
def crear_lista(tamanyo):
    lista = []
    for i in range(tamanyo):
        lista.append(str(i) + '\n')
    return lista

ruta = os.path.join("res", "a_dummy.txt")

# Escritura de un archivo utilizando modo 'wt' (write text)
with open(ruta, 'wt') as fichero:
    fichero.write('Cabecera del ciclo del for en el ejemplo\n')
    lista = crear_lista(10)
    fichero.writelines(lista)
    print("archivo creado")
```

Buffering
* Por defecto, el texto que transfieres desde tu programa a un fichero no se guarda en disco inmediatamente. Se almacena en un buffer.
* Acciones como cerrar un fichero o invocar el método `flush` fuerzan que se transfiera el contenido del buffer a disco.

```python
ruta = os.path.join("res", "FicheroPara Escritura.txt")

# Apertura en modo escritura e inserción de texto en el buffer
fichero_write = open(ruta, 'w')
fichero_write.write('foter')

# Intento de lectura antes de hacer flush/close
fichero_read = open(ruta, "r")
print(fichero_read.readlines())

# Forzamos la escritura en el disco
fichero_write.flush()
print(fichero_read.readlines())

fichero_write.close()
fichero_read.close()
```

## 1.4 Archivos CSV
* Python permite leer datos de ficheros CSV y también escribir ficheros en este formato.
* Popular formato en ciencia de datos.

```python
# tabla_operaciones.csv contiene valores separados por comas
import csv

ruta = os.path.join("res", "tabla_operaciones.csv")

# Lectura básica de archivo CSV
with open(ruta) as fichero:
    data_reader = csv.reader(fichero, delimiter=',')
    for linea in data_reader:
        print(linea[0] + " | " + linea[1])
```

```python
# Leer desde un CSV y escribir en otro nuevo al mismo tiempo
ruta = os.path.join("res", "tabla_operaciones.csv")
f = open(ruta)
data_reader = csv.reader(f, delimiter=',')

ruta_o = os.path.join("res", "4_tabla_operaciones.csv")

with open(ruta_o, 'w') as f2:
    csv_writer = csv.writer(f2, delimiter=',')
    for line in data_reader:
        print(line)
        csv_writer.writerow(line)
        
f.close()
```

# 2 Docstrings
* Python permite adjuntar documentación a los objetos e inspeccionarla a través de la línea de comandos o durante la ejecución del programa.
* Los docstrings se almacenan en el atributo `__doc__` de cada objeto. El valor de dicho atributo se puede mostrar por medio de la función `help`.

```python
# Uso de la función help en comandos nativos
help(open)
```

No es necesario editar este atributo directamente. Para asociar un docstring a un objeto basta con escribir el texto (entre triples comillas) al principio de los modulos, funciones o clases, antes del codigo ejecutable.

```python
# Función con docstring básico
def funcion_de_prueba():
    """Esta es la documentación de la función de prueba"""
    pass

help(funcion_de_prueba)
```

Accediendo al atributo `__doc__` sólo se obtiene el docstring.

```python
# Acceso directo al docstring omitiendo el formato extra de help()
print(funcion_de_prueba.__doc__)
```

Como se puede observar, `help` añade información adicional. Por ejemplo, para entidades más grandes, `help` muestra el docstring dividido en secciones.

```python
import sys
help(sys)
```

Los docstrings que ocupan más de una línea suelen tener estas partes:
* Resumen en la primera línea.
* Línea en blanco.
* Descripción detallada.
* Otra línea en blanco antes del código.

```python
# Función con docstring multilínea
def funcion_de_prueba2():
    """Esta es la línea para el resumen.
    
    Este es el párrafo donde se puede escribir una descripción más
    detallada de la función. Observa que el resumen y la descripción
    detallada van separados por una línea en blanco.
    
    También hay otra línea en blanco antes de empezar el código.
    """
    pass

help(funcion_de_prueba2)
```

Algunos comentarios adicionales:
* Comentarios con '#' suelen asociarse a expresiones sencillas, instrucciones individuales o pequeños bloques de código.
* Los docstrings son más apropiados para construcciones de más alto nivel: funciones, clases y módulos.

### Ejercicio 1: Búsqueda de subcadenas en un fichero
**Enunciado:** Escribe una función que reciba una ruta de un fichero de texto y una cadena de caracteres a buscar y determine si la cadena aparece en el fichero.

```python
import os

def buscar_cadena(ruta_fichero, cadena_a_buscar):
    """
    Busca una cadena específica dentro de un archivo de texto.
    Retorna True si la encuentra, False en caso contrario.
    """
    try:
        # Abrimos el archivo en modo lectura ('r') y con codificación UTF-8
        with open(ruta_fichero, 'r', encoding='utf-8') as fichero:
            contenido = fichero.read()
            
            # Verificamos si la cadena exacta está dentro del texto
            if cadena_a_buscar in contenido:
                return True
            else:
                return False
                
    except FileNotFoundError:
        print(f"Error: El archivo en '{ruta_fichero}' no existe.")
        return False
```

```python
# --- Celda para probar el Ejercicio 1 ---
# Asegúrate de tener un archivo 'prueba.txt' para que esto funcione
ruta_ej1 = os.path.join("prueba.txt")

# Descomenta las siguientes líneas si quieres crear el archivo de prueba rápido:
# with open(ruta_ej1, 'w') as f: f.write("Hola, este es un texto de prueba para Jupyter.")

resultado = buscar_cadena(ruta_ej1, "texto de prueba")
print(f"¿Se encontró la cadena?: {resultado}")
```

### Ejercicio 2: Guardar los primeros n elementos de una lista
**Enunciado:** Escribe una función que reciba una lista, una ruta destino y un número n. La función debe crear un fichero en la ruta especificada. El contenido del fichero serán los primeros n elementos de la lista. La función debe controlar de manera apropiada los posibles valores de n que estén fuera de rango.

```python
def guardar_primeros_elementos(lista, ruta_destino, n):
    """
    Guarda los primeros 'n' elementos de una lista en un archivo de texto.
    Controla que 'n' no sea negativo ni mayor al tamaño real de la lista.
    """
    # Control de rango
    if n < 0:
        print("Error: El número de elementos (n) no puede ser negativo.")
        return 
    
    if n > len(lista):
        print("Aviso: 'n' es mayor que el tamaño de la lista. Se guardará toda la lista completa.")
        n = len(lista)
        
    # Extraemos la porción de la lista que nos interesa
    elementos_a_guardar = lista[:n]
    
    try:
        # Abrimos/creamos el archivo en modo escritura ('w')
        with open(ruta_destino, 'w', encoding='utf-8') as fichero:
            for elemento in elementos_a_guardar:
                fichero.write(str(elemento) + '\n')
                
        print(f"¡Éxito! Se han guardado {n} elementos en '{ruta_destino}'.")
        
    except Exception as error:
        print(f"Ocurrió un error de escritura: {error}")
```

```python
# --- Celda para probar el Ejercicio 2 ---
mi_lista = ["Manzana", "Banana", "Cereza", "Dátil", "Saúco"]
ruta_ej2 = os.path.join("frutas.txt")

# Probamos guardar los primeros 3 elementos
guardar_primeros_elementos(mi_lista, ruta_ej2, 3)
```

### Ejercicio 3: Frecuencia de palabras en un texto
**Enunciado:** Escribe una función que reciba una ruta de un fichero de texto y devuelva un diccionario con la frecuencia de aparición de cada palabra. 
*Ejemplo:* Un fichero que contenga la frase 'es mejor que venga que que no venga' devolverá: `{'es': 1, 'mejor': 1, 'que': 3, 'venga': 2, 'no': 1}`. 
Para dividir un string en palabras puedes hacer uso del método split.

```python
def frecuencia_palabras(ruta_fichero):
    """
    Lee un archivo de texto y devuelve un diccionario indicando 
    cuántas veces se repite cada palabra.
    """
    diccionario_frecuencias = {}
    
    try:
        with open(ruta_fichero, 'r', encoding='utf-8') as fichero:
            # Recorremos el archivo línea por línea
            for linea in fichero:
                # split() divide la línea en palabras eliminando los espacios extra
                palabras = linea.strip().split()
                
                for palabra in palabras:
                    # Pasamos a minúsculas para unificar "Que" y "que"
                    palabra_limpia = palabra.lower()
                    
                    if palabra_limpia in diccionario_frecuencias:
                        diccionario_frecuencias[palabra_limpia] += 1
                    else:
                        diccionario_frecuencias[palabra_limpia] = 1
                        
        return diccionario_frecuencias
        
    except FileNotFoundError:
        print(f"Error: El archivo '{ruta_fichero}' no existe.")
        return {}
```

```python
# --- Celda para probar el Ejercicio 3 ---
ruta_ej3 = os.path.join("frase.txt")

# Creamos un archivo de prueba rápido con el ejemplo del profesor
with open(ruta_ej3, 'w', encoding='utf-8') as f:
    f.write("es mejor que venga que que no venga")

# Ejecutamos la función y mostramos el diccionario resultante
frecuencias = frecuencia_palabras(ruta_ej3)
print(frecuencias)
```
