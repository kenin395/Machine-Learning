## 1.1 Selección (sentencias condicionales)
Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ':'

```python
# Pedimos un dato al usuario. Como lo que escribe llega como texto, usamos int() para convertirlo a número.
x = int(input("Ingrese el valor de x: "))

# Verificamos si 'x' es mayor a 0. Si esto es verdadero, entra al primer bloque.
if x > 0:
    print('Valor positivo')
else:
    # Si la condición de arriba fue falsa (es cero o negativo), el programa cae obligatoriamente aquí.
    print('Valor no positivo')
```

Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]

# Verificamos simplemente si 3 es mayor a 0
print(x > 0) 

# Podemos unir condiciones. Aquí preguntamos si x es mayor a 0 Y además menor a 10
print(x > 0 and x < 10) 

# Comprobamos si las variables no son el mismo objeto en memoria (un número y una lista son diferentes)
print(x is not y) 

# Revisamos si el número 3 está guardado dentro de la lista 'y'
print(x in y)
```

Se pueden introducir más 'ramas' en sentencias condicionales a través de la palabra clave elif.

```python
x = -3

# Evaluamos la primera opción
if x > 0:
    print('Valor positivo')
# Si la primera falló, el 'elif' nos da una segunda oportunidad para evaluar otra cosa
elif x == 0:
    print('Valor nulo')
# Si fallaron tanto el 'if' como el 'elif', por descarte usamos el 'else'
else:
    print('Valor negativo')
```

Anidamiento.

```python
val = 120

# Primero comprobamos si es mayor a cero
if val > 0:
    # Si entró aquí, abrimos un nuevo nivel de condiciones adentro (esto es el anidamiento)
    if val < 100:
        print('Valor positivo')
    else:
        # Esto pasa si el número fue mayor a cero, pero NO fue menor a 100 (es decir, 100 o más)
        print('Valor muy positivo')
else:
    # Este else le pertenece al primer if de arriba de todo
    print('Valor negativo')
```

# 2 Switch

```python
# El 'match' funciona igual que el Switch de otros lenguajes. 
# Toma el valor de 'a' y busca una coincidencia exacta hacia abajo.
a = -6
match a:
    case 1: 
        print('uno')
    case _:
        print('negative')

a = "1"
# Evaluamos a qué día de la semana corresponde la letra
match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        # El guion bajo es el "caso por defecto". Si no coincidió con ninguna letra, salta aquí.
        print("Wrong Day")
```

Expresión ternaria
* Estructura if-else condensado en una línea.
* Se recomienda usar solo en casos sencillos.

```python
# Ejemplo: cálculo del valor absoluto de un número.
x = -3

# Forma de leerlo: el resultado será 'x' si x es mayor o igual a 0. De lo contrario, el resultado será '-x'
resultado = x if x >= 0 else -x
print(resultado)
```

```python
# Aquí hacemos exactamente lo mismo de arriba, pero con la estructura tradicional (más larga)
val = -3
if val >= 0:
    resultado = val
else:
    resultado = -val

print(resultado)
```

Comparación de variables: '==' vs 'is'
* '==' compara si el valor de las variables es el mismo
* 'is' compara si los objetos en las variables son iguales (referencia)

```python
a = 1000
b = a

# 'is' revisa si son literalmente la misma cosa en la memoria de la computadora
print(a is b) 
# '==' revisa si el contenido de adentro vale lo mismo
print(a == b)

c = type(a)
d = type(b)
print(c == d)
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

# 'a' y 'b' tienen los mismos números, pero son dos listas diferentes creadas por separado
print(a is b) # Falso
# 'c' se asignó como un espejo de 'a', así que son el mismo objeto exacto
print(a is c) # Verdadero

print(a == b) # Verdadero, porque los números que tienen dentro son idénticos
print(id(a))
print(id(b))
print(id(c))
```

Valor booleano
Todos los objetos en Python tiene inherentemente un valor booleano: True o False.
* Cualquier número distinto de 0 ó cualquier objeto no vacío tienen valor True.
* El número 0, objetos vacíos y el objeto especial None tienen valor False.

```python
a = 10    # Cualquier número que no sea cero, Python lo considera True
b = None  # None (nada) se considera False

if a:
    # Va a imprimir esto porque 'a' tiene un valor válido
    print("a es True")
if not b:
    # Usamos 'not' para invertirlo. Como b es False, 'not b' se vuelve True y entra.
    print("b es False")
```

```python
c = [1, 2, 3]

# Vaciamos la variable asignándole None. 
c = None

# Si hacemos un if de una variable vacía, automáticamente da falso.
if c:
    print("Lista no vacía")
else:
    print("Lista vacía")
```

# 3 Iteración (bucles)
* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle. Dos tipos: while y for.

Bucle 'while'
* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a False).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.

```python
# Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]

# El bucle va a girar mientras el índice siga siendo menor a 3 (0, 1, 2)
while indice < 3:
    print(numeros[indice])
    # Regla de oro del while: siempre cambiar el valor para que el bucle avance, si no se queda colgado
    indice = indice + 1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0

# Iteramos mientras no nos salgamos del tamaño total de la lista
while indice < len(numeros):
    # Si el número que estamos viendo ahorita es menor a 10...
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1 # Le sumamos uno a nuestro contador de números pequeños
    
    indice += 1 # Pasamos a revisar el siguiente número de la lista

print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'

# Mientras la variable 'nombre' no se quede vacía, esto va a seguir repitiéndose
while nombre: 
    print(nombre)
    # Le cortamos la primera letra al nombre en cada vuelta usando slicing
    nombre = nombre[1:]
```

```python
# OJO: Bucle infinito (hemos comentado el aumento de i para que no te cuelgue el Jupyter si te equivocas)
i = 0
while i < 10:
    print(i)
    # Si comentas esta línea de abajo, la 'i' siempre valdrá 0, nunca llegará a 10, y el bucle jamás se detendrá.
    i += 1
```

Bucle 'for'
* Permite recorrer los items de una sequencia o un objeto iterable. Funciona en strings, listas, tuplas, etc.

```python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']

# El for es automático. En cada vuelta saca una película de la lista y la mete en la variable temporal 'pelicula'
for pelicula in peliculas:
    print(pelicula)
```

También se usa para iterar un número preestablecido de veces (counted loops):

```python
# range(10) genera los números del 0 al 9. El for dará exactamente 10 vueltas.
for i in range(10):
    print(i)
```

range es útil en combinación con len porque permite acceder a los elementos de una secuencia por posición.

```python
nombre = 'Pablo'

# len saca que Pablo tiene 5 letras. range crea números del 0 al 4. El for los usa como posiciones.
for i in range(len(nombre)):
    print(nombre[i])
```

```python
# Esta es una forma mucho más directa y recomendada de hacer lo mismo que arriba.
for letra in nombre:
    print(letra)
```

Iteración de tuplas:

```python
tuplas = [(1, 2, 4), (3, 4, 0), (5, 6, 0)] # Se ajusto la lista para evitar el error del PDF (todas deben tener 3 datos)

# Como cada elemento de la lista es un paquete de 3 datos, los desempaquetamos directo en a, b y c
for a, b, c in tuplas:
    print(a, b, c)
```

Iteración de diccionarios. Se iteran las claves:

```python
diccionario = {'a': 1, 'b': 2, 'c': 3}

# Si iteramos el diccionario así nomás, Python solo nos da las letras (las claves)
for key in diccionario:
    print(f"{key} => {diccionario[key]}")
```

* Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos items y values.

```python
# items() nos da el paquete completo. En cada vuelta recibimos la letra en 'key' y el número en 'value'
for key, value in diccionario.items():
    print(f"{key} => {value}")
```

```python
# Si solo nos importan los números (los valores), usamos values()
for value in diccionario.values():
    print(value)
```

La variable item en la cabecera del for puede ser cualquier expresión que sea válida como parte izquierda de una asignación convencional.

```python
# Python es muy flexible y permite extraer variables al vuelo directamente desde la cabecera del for
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
```

## 3.1 Las sentencias break, continue y else
Break y continue
* Solo tienen sentido dentro de bucles.
* Break permite terminar el bucle por completo.
* Continue permite saltar a la siguiente iteración, continuando con el bucle.
* Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (if).

```python
# Ejemplo: break
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]

for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        # El break destruye el bucle al instante. Así nos ahorramos procesar el resto de la lista.
        break
```

```python
# Ejemplo: continue
for i in range(10):
    # Si el número que estamos revisando dividido para 2 sobra 0 (es decir, es par)...
    if i % 2 == 0:
        # El continue hace que el bucle ignore lo que falta del código abajo y salte a la próxima vuelta
        continue 
    print(f'Odd number {i}')
```

Observa como la sentencia continue te puede ayudar a reducir el número de niveles de anidamiento.
Sin continue el anterior ejemplo sería:

```python
# Al no usar continue, tenemos que anidar el print dentro del if, lo cual hace el código más engorroso si es muy largo
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
```

```python
# El break solo rompe el bucle en el que está metido directamente (el bucle de la variable 'a').
for i in range(5):
    for a in range(2):
        print(f"{i}-{a}")
        if i == 3 and a == 0:
            break
```

Else
Los bucles pueden tener una sensencia else.
* Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.
* Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un break.

```python
lista = [60, 60, 30, 60]
element_to_find = 60

for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    # IMPORTANTE: Este else le pertenece al FOR. Solo se imprimirá si nunca se ejecutó el break de arriba
    print("not found")
```

```python
# with flag (usar una variable bandera para saber si encontramos algo)
lista = [10, 30, 50, 40]
element_to_find = 30

# Empezamos asumiendo que no lo hemos encontrado
found = False

for element in lista:
    if element == element_to_find:
        found = True # Si lo encontramos, cambiamos la bandera a Verdadero
        break

# Al final del bucle, revisamos cómo quedó la bandera
if found:
    print("dato encontrado")
else:
    print("not found")
```

# 4 List Comprehensions
* Permiten construir listas a través de la ejecución repetida (sentencia for) de una expressión para cada item de un objeto iterable.
* Van entre '[' y ']'. Esto es indicativo de que estamos construyendo una lista.
Sintaxis:
[<expression> for <item> in <iterable>]
Ejemplo: repetir los carácteres de un string.

```python
# En una sola línea, toma la palabra Enrique, saca cada letra y la repite 3 veces
letras = [char*3 for char in "Enrique"]
print(letras)
```

```python
# Esta es la forma larga y antigua de hacer exactamente lo mismo que hicimos arriba
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
```

Comúnmente, el item de la sentencia for aparecerá en la expresión principal, pero eso no es obligatorio.

```python
# Generamos un número 2 por cada letra que haya en "Enrique" (son 7 letras, así que genera 7 dos).
numeros = [2 for x in 'Enrique']
print(numeros)
```

### 4.0.1 Versión extendida
Se puede especificar un filtro (sentencia if) para obtener únicamente los elementos que cumplan cierta condición.
Sintaxis:
[<expression> for <item> in <iterable> if <condition>]
Ejemplo: obtener números pares:

```python
# Crea un número (y) del 0 al 8, pero al final tiene un filtro if: solo lo guarda si es par
pares = [y for y in range(9) if y % 2 == 0]
print(pares)
```

Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle equivalente.

```python
# Versión sin comprehension (más líneas de código pero hace lo mismo de arriba)
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
```

### 4.0.2 Versión completa
La sentencia if de una list comprehension también puede contener una expresión alternativa.
Sintaxis:
[<expression_1> if <condition> else <expression_2> for <item> in <iterable>]
Ejemplo: poner a cero los números pares.

```python
# Se lee así: Pon un 0 si el número es par, caso contrario deja la x original. Haz esto para números del 0 al 8.
numeros_mixtos = [0 if x % 2 == 0 else x for x in range(9)]
print(numeros_mixtos)
```

```python
# Eleva al cuadrado si es mayor a 2, sino déjalo igual... pero solo si es positivo (x>0)
resultado = [x**2 if x > 2 else x for x in range(-4, 5) if x > 0]
print(resultado)
```

```python
# Equivalente clásico a la comprehension de arriba
lista = []
for x in range(-4, 5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
print(lista)
```

```python
# Se pueden encadenar múltiples condiciones (como un if-elif-else) dentro de la comprehension
resultado = [x**2 if x > 0 else 100 if x == 0 else x for x in range(-4, 5)]
print(resultado)
```

```python
# El equivalente largo a la celda anterior
lista = []
for x in range(-4, 5):
    if x > 0:
        lista.append(x**2)
    elif x == 0:
        lista.append(100)
    else:
        lista.append(x)
print(lista)
```

### 4.0.3 Anidamiento
Las list comprehensions soportan anidamiento en sus expresiones.

Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas.

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]

# Mezcla las dos listas. Por cada elemento de lista_1, recorre toda la lista_2 y crea parejas
parejas = [(x, y) for x in lista_1 for y in lista_2]
print(parejas)
```

### 4.0.4 Pros y contras
* Principales ventajas de las comprehensions en comparación a un bucle convencional:
  * Expresión compacta y legible, si estás familiarizado con la sintaxis.
  * Mejor rendimiento.
* Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una expresión difícil de entender.

### 4.0.5 Otros tipos de comprehensions
Dictionary comprehensions
Usando '{' y '}' como delimitadores y una expresión 'clave: valor', se obtiene un diccionario en lugar de una lista.
Ejemplo: diccionario donde cada valor es el cuadrado de la clave.

```python
# Notar los dos puntos (:), eso le dice a Python que estamos armando un diccionario. Clave es x, Valor es x*x.
d = {x: x*x for x in range(10)}
print(d)
print(type(d))
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]

# zip une dos listas por posición. Aquí creamos un diccionario uniendo el nombre (k) con el precio (v)
shopping = {k: v for k, v in zip(items, price)}
print(shopping)
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]

# Esto es para que veas que la función zip por debajo nos entrega pequeñas tuplas (parejas) en cada vuelta
for k in zip(items, price):
    print(type(k))
    print(k)
```

Set comprehensions
Usando '{' y '}' como delimitadores y una expresión simple (al igual que en las list comprehensions), se obtiene un conjunto.
Ejemplo: conjunto que incluye los 10 primeros números naturales.

```python
# Un Set (conjunto) usa llaves pero sin los dos puntos que usan los diccionarios
c = {x for x in range(10)}
print(c)
print(type(c))
```

Tuple comprehensions

```python
# Para las tuplas se usa la palabra tuple() rodeando la comprehension. (Los paréntesis solos crean un generador)
tupla = tuple(x for x in range(4))
print(tupla)
```

# 5 Excepciones
* Las excepciones son eventos que representan situaciones excepcionales. Alteran el flujo de ejecución convencional.
* Python lanza excepciones automáticamente cuando se producen errores.
* El programador puede lanzar excepciones de manera explícita y también capturar excepciones para actuar como se crea conveniente.

Try/except
* Permite capturar excepciones y actuar en consecuencia.
Ejemplo: error de acceso fuera de rango.

```python
lista = [6, 1, 0, 5]
i = 300

# Usamos Try para envolver código peligroso. Si falla, en lugar de cerrarse la consola, salta al Except.
try:
    # Intentamos acceder a la posición 300 (la lista solo llega al 3, así que esto explotará)
    a = lista[i]
except IndexError:
    # Como falló por pedir un índice que no existe (IndexError), el programa se salva aquí
    print('He capturado la excepción de tipo IndexError')

a = 0
```

* Normalmente, al capturar excepciones, querremos ser lo más específicos posible, pero también se puedes usar una sentencia try-except que capture cualquier error.

```python
# Si ponemos except sin nada más, va a atrapar cualquier tipo de error sea el que sea
try:
    4/0
except:
    print('He capturado la división por cero')
```

* Recoger la excepción e imprimir el mensaje de error

```python
lista = [6, 1, 0, 5]

try:
    print(lista[4])
except Exception as e:
    # Si queremos guardar el mensaje de error original de Python para leerlo, usamos 'as e'
    print("Error = " + str(e))
```

Try/finally
A través de finally podemos especificar código que queremos que se ejecute siempre (independientemente de si se produce la excepción o no).
Se suele usar para liberar recursos.

```python
lista = [6, 1, 0, 5]

try:
    a = lista[1]
except IndexError:
    print('Exception IndexError Captured')
finally:
    # Este print se ejecutará pase lo que pase, así no haya existido ningún error arriba
    print('Bloque finally')
```

raise
* La sentencia raise nos permite lanzar excepciones de manera explícita.

```python
lista = [6, 1, 0, 5]
indice = 4

try:
    if indice >= len(lista):
        # Con raise somos nosotros quienes provocamos el error a propósito y le ponemos un mensaje personalizado
        raise IndexError('Índice fuera de rango')
except IndexError:
    print('Excepción de tipo IndexError capturada')
```

```python
# Ejercicio 1
numeros = [1, 5, 9, 2, 3]
suma = 0

for numero in numeros:
    suma += numero # En cada iteración acumulamos el valor de la lista

print(suma)
```

```python
# Ejercicio 2
numeros = [1, 2, 2, 3, 4, 5, 4, 5, 3, 3, 1]
resultado = []

for numero in numeros:
    # Revisamos si el número de la vuelta actual aún no se ha guardado en la nueva lista
    if numero not in resultado:
        resultado.append(numero)

print(resultado)
```

```python
# Ejercicio 3
n = 5
diccionario = {}

# range(1, n+1) genera del 1 al 5. Asignamos como valor el cuadrado del iterador.
for i in range(1, n+1):
    diccionario[i] = i*i

print(diccionario)
```

```python
# Ejercicio 4
palabras = ["", "Valencia", "python", "asignaturas", "programación", "java", "estudiantes", "académicos"]

for palabra in palabras:
    # Comprobamos que el tamaño sea mayor o igual a 9 y que la letra en posición 0 sea una 'a'
    if len(palabra) >= 9 and palabra[0] == 'a':
        print(f"Palabra encontrada: {palabra}")
        break # Abortamos apenas la encontramos
else:
    # Este else solo se acciona si el for terminó sin encontrar nada
    print("Palabra no encontrada")
```

```python
# Ejercicio 5
L = [3, 5, 13, 12, 1, 9]
resultado = []

for i in range(len(L)):
    # Modulo 3. Si el residuo es 0, significa que el número en esa posición es múltiplo de 3
    if L[i] % 3 == 0:
        resultado.append(i) # Guardamos el índice (la posición)

print(resultado)
```

```python
# Ejercicio 6
d = {'a': 4.3, 'b': 1, 'c': 7.8, 'd': -5}
maximo = float("-inf") # Empezamos con negativo infinito como valor base para comparar

if len(d) == 0:
    print("El diccionario está vacío")

# Recorremos el diccionario llave a llave y comparamos el valor actual con nuestro 'maximo' guardado
for clave, valor in d.items():
    if valor > maximo:
        maximo = valor
        resultado = clave

print(resultado)
```

```python
# Ejercicio 7
a = [2, 4, 6, 8]
b = [7, 11, 15, 22]

for i in a:
    if i > 5: # Validamos la regla para la lista a
        for j in b:
            if j < 14: # Validamos la regla para la lista b
                print(f"{i} x {j} = {i*j}")
```

```python
# Ejercicio 8
try:
    x = input("Ingresa un número: ")
    numero = float(x) # Si escribió texto, esta línea revienta y se va al ValueError
    print(f"El resultado de 10/{x} es: {10/numero}") # Si puso un 0, esta línea revienta al ZeroDivisionError
except ValueError:
    print(f"El valor '{x}' no es numérico")
except ZeroDivisionError:
    print("Error: división por cero")
```

```python
# Ejercicio 9
notas = {'Programación': 9.5, "Inteligencia artificial": 7.6, "Redes": 5.2, "Sistemas operativos": 6.4}

try:
    asignatura = input("Introduce una asignatura: ")
    print(f"La nota de la asignatura '{asignatura}' es: {notas[asignatura]}")
except KeyError:
    # Si consultamos el diccionario con una clave que no existe, nos lanza el KeyError y lo atajamos
    print(f"Asignatura incorrecta: {asignatura}")
```

```python
# Ejercicio 10
lista = [1, 4, -3, -1.5, 6.5, 2, 8, 2.1]

# Filtramos pidiendo que el tipo de dato sea int (entero) y que sea positivo (>0)
resultado = [numero for numero in lista if type(numero) == int and numero > 0]
print(resultado)
```

```python
# Ejercicio 11
palabra = "murcielago"

# Usamos llaves {} para crear un Set, que automáticamente elimina vocales duplicadas
vocales = {letra for letra in palabra if letra in ('a', 'e', 'i', 'o', 'u')}
print(vocales)
```

```python
# Ejercicio 12
# Iteramos del 0 al 50. Convertimos a string cada número para revisar si tiene un '3' escrito adentro
numeros_tres = [numero for numero in range(51) if '3' in str(numero)]
print(numeros_tres)
```

```python
# Ejercicio 13
frase = "Soy un ser humano"

# .split() parte la oración en palabras. Guardamos la palabra como clave y su longitud (len) como valor
tamano_palabras = {palabra: len(palabra) for palabra in frase.split()}
print(tamano_palabras)
```

```python
# Ejercicio 14
numero_a_palabra = {6: "seis", 7: "siete", 8: "ocho", 9: "nueve", 10: "diez"}

# Corregido: Si el iterador es menor O IGUAL a 5 (<= 5), guarda su número real. 
mixtos = [numero if numero <= 5 else numero_a_palabra[numero] for numero in range(1, 11)]
print(mixtos)
```
