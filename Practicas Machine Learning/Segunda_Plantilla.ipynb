# 2 Literales, variables y tipos de datos básicos

De forma muy genérica, al ejecutarse un programa Python, simplemente se realizan operaciones sobre objetos. Estos dos términos son fundamentales.
* **Objetos**: cualquier tipo de datos (números, caracteres o datos más complejos).
* **Operaciones**: cómo manipulamos estos datos.

```python
4 + 3
```

## 2.1 Literales
* Python tiene una serie de tipos de datos integrados en el propio lenguaje.
* Los literales son expresiones que generan objetos de estos tipos.
* Estos objetos, según su tipo, pueden ser:
  * Simples o compuestos.
  * Mutables o inmutables.

**Literales simples**
* Enteros
* Decimales o punto flotante
* Booleano

```python
print(4)          # número entero
print(4.2)        # número en coma flotante
print('Hello world!') # string
print(False)      # booleano
```

**Literales compuestos**
* Tuplas
* Listas
* Diccionarios
* Conjuntos

```python
print([1, 2, 3, 3]) # lista (mutable)
print({'Nombre': 'John Doe', "edad": 30}) # Diccionario (mutable)
print({1, 2, 3, 3}) # Conjunto (mutable)
print((4, 5)) # tupla (inmutable)
```

## 2.2 Variables

**Referencias a objetos.**
* Las variables y los objetos se almacenan en diferentes zonas de memoria.
* Las variables siempre referencian a objetos y nunca a otras variables.
* Objetos sí que pueden referenciar a otros objetos. Ejemplo: listas.
* Sentencia de asignación: `<nombre_variable> = <objeto>`

```python
# Asignación de variables
a = 5
print(a)
```

```python
a = 1         # entero
b = 4.0       # coma flotante
c = "ITQ"     # string
d = 10 + 1j   # numero complejo
e = True      # boolean
f = None      # None

# visualizar valor de las variables y su tipo
print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
print(e, type(e))
print(f, type(f))
```

**Las variables no tienen tipo.**
* Las variables apuntan a objetos que sí lo tienen.
* Dado que Python es un lenguaje de tipado dinámico, la misma variable puede apuntar, en momentos diferentes de la ejecución del programa, a objetos de diferente tipo.

```python
a = 3
print(a, type(a))

a = 'Pablo García'
print(a, type(a))

a = 4.5
print(a, type(a))
```

* **Garbage collection:** Cuando un objeto deja de estar referenciado, se elimina automáticamente.

**Identificadores**
Podemos obtener un identificador único para los objetos referenciados por variables.
* Este identificador se obtiene a partir de la dirección de memoria.

```python
a = 3
print(id(a))

a = 'Pablo García'
print(id(a))

a = 4.5
print(id(a))
```

* **Referencias compartidas:** un mismo objeto puede ser referenciado por más de una variable. Variables que referencian al mismo objeto tienen mismo identificador.

```python
a = 4567
print(id(a))

b = a
print(id(b))

c = 4567
print(id(c))
```

```python
a = 25
b = 25
print(id(a))
print(id(b))
print(id(25))
```

```python
# Ojo con los enteros "grandes" [-5, 256]
a = 258
b = 258
print(id(a))
print(id(b))
print(id(258))
```

```python
# Referencia al mismo objeto a través de asignar una variable a otra.
a = 400
b = a
print(id(a))
print(id(b))
```

**Las variables pueden aparecer en expresiones.**

```python
a = 3
b = 5
print(a + b)

c = a + b
print(c)
print(id(c))
```

**Respecto a los nombres de las variables**
* No se puede poner números delante del nombre de las variables.
* Por convención, evitar CamelCase. Mejor usar snake_case: uso de `_` para separar palabras.
* El lenguaje diferencia entre mayúsculas y minúsculas.
* Deben ser descriptivos.
* Hay palabras o métodos reservados -> Built-ins y KeyWords
* ¡Ojo con reasignar un nombre reservado!

```python
print(pow(3, 2))

# Ejemplo de error al reasignar built-in (descomentar para probar)
# pow = 1 
# print(pow)
# print(pow(3, 2)) # Dará TypeError
```

**Asignación múltiple de variables**

```python
x, y, z = 1, 2, 3
print(x, y, z)

t = x, y, z, 7, "Python"
print(t)
print(type(t))
```

Esta técnica tiene un uso interesante: el intercambio de valores entre dos variables.

```python
a = 1
b = 2
a, b = b, a
print(a, b)
```

```python
a = 1
b = 2
c = a
a = b
b = c
print(a, b, c)
```

## 2.3 Tipos de datos básicos

**Bool**
* 2 posibles valores: `True` o `False`.

```python
a = False
b = True
print(a, type(a))
print(b, type(b))
```

`True` y `False` también son objetos que se guardan en caché, al igual que los enteros pequeños.

```python
a = True
b = True
print(id(a))
print(id(b))
print(a is b)
print(a == b)
```

**Números**
* Diferentes representaciones: base 10, 2, 8, 16.

```python
print(2)      # Enteros, sin parte fraccional.
print(3.4)    # Números en coma flotante, con parte fraccional.
print(2+4j)   # Números complejos.
print(1/2)    # Numeros racionales.
```

```python
x = 58          # decimal
z = 0b00111010  # binario
w = 0o72        # octal (0o en lugar de 00)
y = 0x3A        # hexadecimal

print(x == y == z == w)
```

**Strings**
Cadenas de caracteres.
* Son secuencias: la posición de los caracteres es importante.
* Son inmutables: las operaciones sobre strings no cambian el string original.

```python
s = 'John "ee" Doe'
print(s[0])       # Primer carácter del string.
print(s[-1])      # Último carácter del string.
print(s[1:8:2])   # Substring ('slicing').
print(s[:])       # Todo el string.
print(s + "e")    # Concatenación.
```

## 2.4 Conversión entre tipos
A veces queremos que un objeto sea de un tipo específico. Podemos obtener objetos de un tipo a partir de objetos de un tipo diferente (casting).

```python
a = int(2.8)
b = int("3")
c = float(1)
d = float("3")
e = str(2)
f = str(3.0)
g = bool("a")
h = bool("")
i = bool(3)
j = bool(0)
k = bool(None)

print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
print(e, type(e))
print(f, type(f))
print(g, type(g))
print(h, type(h))
print(i, type(i))
print(j, type(j))
print(k, type(k))
```

```python
print(7/4)      # División convencional. Resultado 'float'
print(7//4)     # División entera. Resultado 'int'
print(int(7/4)) # Divisón convencional y conversión a 'int'
```

## 2.5 Operadores
Python3 precedencia en operaciones
* Combinación de valores, variables y operadores
* Operadores y operandos

**Operadores aritméticos**
* `+` Suma
* `-` Resta
* `/` División
* `//` División Entera
* `%` Modulo/Resto
* `*` Multiplicacion
* `**` Exponenciación

```python
x = 3
y = 2
print('x + y =', x + y)
print('x - y =', x - y)
print('x * y =', x * y)
print('x / y =', x / y)
print('x // y =', x // y)
print('x % y =', x % y)
print('x ** y =', x ** y)
```

**Operadores de comparación**
* `>` Mayor
* `<` Menor
* `==` Igualdad
* `!=` Desigualdad
* `>=` Mayor o Igual
* `<=` Menor o Igual

```python
x = 10
y = 12
print('x > y es', x > y)
print('x < y es', x < y)
print('x == y es', x == y)
print('x != y es', x != y)
print('x >= y es', x >= y)
print('x <= y es', x <= y)
```

**Operadores Lógicos**
* `and` : True, si ambos son True
* `or` : True, si alguno de los dos es True
* `not` : Negación

```python
x = True
y = False
print('x and y es :', x and y)
print('x or y es :', x or y)
print('x xor y es :', x != y) # XOR en python es != para booleanos
print('not x es :', not x)
```

**Operadores de Identidad**
* `is` : True, si ambos operadores son una referencia al mismo objeto.
* `is not` : True, si ambos operadores no son referencia al mismo objeto.

```python
a = 4444
b = a
print(a is b)
print(a is not b)
```

**Operadores de Pertenencia**
* `in` : True, si a se encuentra en la secuencia b
* `not in` : True, si a no se encuentra en la secuencia b

```python
x = 'Hola Mundo'
y = {1: 'a', 2: 'b'}

print('H' in x)
print('hola' not in x)
print(1 in y)
print('a' in y)
```

# 3 Tipos de datos compuestos (colecciones)
## 3.1 Listas
Una colección de objetos.
* Mutables.
* Tipos arbitrarios heterogeneos.
* Puede contener duplicados.
* No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.
* Los elementos van ordenados por posición. Se acceden usando la sintaxis: `[index]`.
* Los índices van de 0 a n-1.
* Soportan anidamiento. Son Arrays Dinámicos.

**Creación de listas.**

```python
letras = ['a', 'b', 'c', 'd']
palabras = 'Hola mundo como estas'.split()
numeros = list(range(7))

print(letras)
print(palabras)
print(numeros, type(numeros))
```

```python
# Pueden contener elementos arbitrarios / heterogeneos
mezcla = [1, 3.4, 'a', None, False]
print(mezcla)
print(len(mezcla)) # len da el tamaño de la lista
```

```python
# Pueden incluso contener objetos más "complejos" (como funciones)
lista_con_funcion = [1, 2, len, pow]
print(lista_con_funcion)
```

```python
# Pueden contener duplicados
lista_con_duplicados = [1, 2, 3, 3, 3, 4]
print(lista_con_duplicados)
```

**Slicing e índices**

```python
letras = ['a', 'b', 'c', 'd', 'e']
print(letras[2])
print(letras[-5])
print(letras[0])

print(letras[1:3])
print(letras[:3])
print(letras[:-1])
print(letras[2:])
print(letras[:])
print(letras[::2])
```

**Añadir e Insertar elementos**

```python
letras = ['a', 'b', 'c', 'd']
letras.append('e')
print(letras)

letras += ['e']
print(letras)

letras.insert(1, 'g')
print(letras)
```

**Modificación de la lista y Slicing múltiple**

```python
letras = ['a', 'b', 'c', 'f', 'g', 'h', 'i', 'j', 'k']
letras[0:7:2] = ['z', 'x', 'y', 'p']
print(letras)
```

```python
# Modificación individual de lista crea anidamiento
numeros = [1, 2, 3]
numeros[1] = [10, 20, 30]
print(numeros)
print(numeros[1][0])

numeros[1][2] = [100, 200]
print(numeros)
```

**Eliminar un elemento**

```python
letras = ['b', 'x', 'y', 'h', 'p', 'j', 'k']
if 'p' in letras:
    letras.remove('p')
print(letras)

elemento = letras.pop() # elimina el elemento al final
print(elemento)
print(letras)
```

**Encontrar índice y Concatenar**

```python
letras = ['a', 'b', 'c', 'c']
if 'a' in letras:
    print(letras.index('a'))
```

```python
lacteos = ['queso', 'leche']
frutas = ['naranja', 'manzana']

compra = lacteos + frutas
print(compra)

# Concatenación sin crear una nueva lista
frutas.extend(['pera', 'uvas'])
print(frutas)

# Anidar sin crear una nueva lista
frutas = ['naranja', 'manzana']
frutas.append(['pera', 'uvas'])
print(frutas)
```

**Replicar y Copiar una lista**

```python
lacteos = ['queso', 'leche']
print(lacteos * 3)

a = 3 * lacteos
print(a)
```

```python
frutas = ['naranja', 'manzana', ['pera', 'uvas']]
frutas2 = frutas.copy() # También puede ser frutas[:]

print(frutas2)
print('id frutas = ' + str(id(frutas)))
print('id frutas2 = ' + str(id(frutas2)))
```
